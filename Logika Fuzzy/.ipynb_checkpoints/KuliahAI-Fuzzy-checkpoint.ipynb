```python
#definisi rentang
x_gaji = np.arange(0, 500000, 75000)
x_masakerja = np.arange(0, 17, 2)
x_umur = np.arange(30, 50, 5)
```

```python
#membuat fungsi keanggotaan untuk gaji
umur_lo = fuzz.trimf(x_umur, [0, 0, 24])
umur_md = fuzz.trimf(x_umur, [0, 24, 48])
umur_hi = fuzz.trimf(x_umur, [24, 48, 48])
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))
ax0.plot(x_umur, umur_lo, 'b', linewidth=1.5, label='Bad')
ax0.plot(x_umur, umur_md, 'g', linewidth=1.5, label='Decent')
ax0.plot(x_umur, umur_hi, 'r', linewidth=1.5, label='Great')
ax0.set_title('Umur Fungsi')
ax0.legend()
```
